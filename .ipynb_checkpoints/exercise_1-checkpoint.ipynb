```python
import numpy as np
from qiskit import QuantumCircuit, Aer, transpile
from math import gcd
from fractions import Fraction

from qiskit.quantum_info import DensityMatrix, partial_trace, hellinger_fidelity
from qiskit.visualization import plot_histogram
```

$\newcommand{\Ket}[1]{\left|#1\right>}$

Shor algo for $N=21, a=4$

```python
N = 21
a = 4
```

Consider $(a^j \ \text{mod} \ N)  \in \{1, 4, 16\} \ \forall_{j \in \mathbb{N}}$.
Therefore, only 3 qubits needed for bottom register: 1, 4, 16
For top register let's choose 4 qubits.

```python
N_QUBITS_TOP = 4
N_QUBITS_BOTTOM = 3
```

We define the operator $U: \Ket{x} \to \Ket{x*4^j \ \text{mod} \ N}$.

```python
def c_4mod21(power):
    """Controlled multiplication by 4^power mod 21 on qubits 1, 4, 16"""
    U = QuantumCircuit(N_QUBITS_BOTTOM)
    if power % 3 == 1:
        U.swap(1, 2)
        U.swap(0, 1)
    elif power % 3 == 2:
        U.swap(0, 1)
        U.swap(1, 2)

    g_U = U.to_gate()
    g_U.name = f"U^{power}: {4**power % N} mod 21"

    c_U = g_U.control()
    return c_U, U
```

We define the inverse QFT on $n$ Qubits:

```python
def qft_dagger(n):
    """n-qubit QFTdagger the first n qubits in circ"""
    qc = QuantumCircuit(n)
    # Don't forget the Swaps!
    for qubit in range(n//2):
        qc.swap(qubit, n-qubit-1)
    for j in range(n):
        for m in range(j):
            qc.cp(-np.pi/float(2**(j-m)), m, j)
        qc.h(j)
    qc.name = "QFT†"
    return qc
```

From this we create the QPE circuit:

```python
def qpe_4mod21_circ(measure=True):
    qc = QuantumCircuit(N_QUBITS_BOTTOM + N_QUBITS_TOP, N_QUBITS_TOP)
    for q in range(N_QUBITS_TOP):
        qc.h(q)  # Initialize counting qubits in state |+>
    qc.x(N_QUBITS_TOP)  # And auxiliary register in state |1>
    for q in range(N_QUBITS_TOP):  # Do controlled-U operations
        c_U, U = c_4mod21(2 ** q)
        qc.append(c_U, [q] + [i + N_QUBITS_TOP for i in range(N_QUBITS_BOTTOM)])
    qc.append(qft_dagger(N_QUBITS_TOP), range(N_QUBITS_TOP))  # Do inverse-QFT
    if measure: qc.measure(range(N_QUBITS_TOP), range(N_QUBITS_TOP))
    return qc
```

Simulate this circuit to extract $r$ until we find prime factors:

```python
def qpe_4mod21():
    """Performs quantum phase estimation on the operation 4*r mod 21.
    Returns:
        float: Estimate of the phase
    """
    qc = qpe_4mod21_circ()
    # Simulate Results
    aer_sim = Aer.get_backend('aer_simulator')
    # `memory=True` tells the backend to save each measurement in a list
    job = aer_sim.run(transpile(qc, aer_sim), shots=1, memory=True)
    readings = job.result().get_memory()
    print("Register Reading: " + readings[0])
    phase = int(readings[0], 2) / (2 ** N_QUBITS_TOP)
    print(f"Corresponding Phase: {phase}")
    return phase
```

```python
for ATTEMPT in range(1, 10): # limited number of attempts
    print(f"\nATTEMPT {ATTEMPT}:")
    phase = qpe_4mod21()  # Phase = s/r
    frac = Fraction(phase).limit_denominator(N)
    r = frac.denominator
    print(f"Frac: {frac}, Result: r = {r}")
    if phase != 0:
        # Guesses for factors are gcd(x^{r/2} ±1 , 21)
        guesses = [gcd(a ** (r // 2) - 1, N), gcd(a ** (r // 2) + 1, N)]
        print(f"Guessed Factors: {guesses[0]} and {guesses[1]}")
        for guess in guesses:
            if guess not in [1, N] and (N % guess) == 0:
                # Guess is a factor!
                print(f"*** Non-trivial factor found: {guess} ***")
                break
        else:
            continue
        break
```

So, the algorithm works in simulation. Let's draw the circuit and get simulated statistics:

```python
qc = qpe_4mod21_circ()
qc.draw("mpl")
```

```python
# Simulate Results
aer_sim = Aer.get_backend('aer_simulator')
job = aer_sim.run(transpile(qc, aer_sim), shots=1024, memory=True)
counts = job.result().get_counts()
plot_histogram(counts)
```

We see peaks for phases $0$, $1/3$ and $2/3$.

Let's run the whole algorithm on the device:

```python
qc_shor = qpe_4mod21_circ()
trans_shor = transpile(qc_shor, device, optimization_level=3)
print("Gates after optimization: ", trans_shor.count_ops())
trans_shor.draw("mpl", fold=100)
```

The transpiled circuit for the 7 qubit IBM machine has 288 CX gates.

Now let's look at the density matrix of the measuring qubits:

```python
m = DensityMatrix(qpe_4mod21_circ(measure=False))
print(m.dims(), m.purity())
m_a = partial_trace(state=m, qargs=[4, 5, 6])
print(m_a.dims(), m_a.purity())
m_a.draw("city")
```

The full 7-qubit density matrix has purity of almost $1$, the density matrix of just the measured qubits has purity around $1/3$, it is entangled.
