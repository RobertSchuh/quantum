```python
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
from tqdm.contrib import tzip
from quantum_tester import TesterGroup, test_on_random_circuits

tg = TesterGroup()
backends = ["ibmq_quito", "ibm_perth", "ibmq_lima", "qb-nm1", "qb-nm2", "qb-qdk1"]
```

```python
nc, shots = 25, 1000
```

```python
ngs = 25 * np.arange(11)
nqs = 2 + 0*ngs

results = [test_on_random_circuits(tg, backends, nq=nq, ng=ng, nc=nc, shots=shots, naive=True) for (nq, ng) in tzip(nqs, ngs)]

for backend in backends:
    partial_res = np.array([result[backend] for result in results])
    plt.errorbar(ngs, np.mean(partial_res, axis=1), yerr=np.std(partial_res, axis=1), label=backend, marker="x", capsize=3)
    # plt.fill_between(ngs, np.min(partial_res, axis=1), np.max(partial_res, axis=1), alpha=0.1)

plt.grid()
plt.legend()
plt.ylabel("Error Score")
plt.xlabel("num gates")
plt.title("2 qubits")
plt.show()
```
